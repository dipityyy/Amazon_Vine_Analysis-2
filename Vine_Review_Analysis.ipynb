```python
import pandas as pd
import numpy as np
```

```python
vine_df = pd.read_csv('vine_table.csv')
vine_df
```

```python
vine_f1 = vine_df[(vine_df["total_votes"] >= 20)]
vine_f1
```

```python
vine_f2 = vine_f1[(vine_f1['helpful_votes']) / (vine_f1['total_votes']) >= 0.5]
vine_f2
```

```python
vine_yes = vine_f2[(vine_f2['vine'] == 'Y')]
vine_yes.count()
```

```python
vine_no = vine_f2[(vine_f2['vine'] == "N")]
vine_no
```

```python
vine_yes_total = vine_yes['review_id'].count()
vine_yes_5star = vine_yes[(vine_yes['star_rating'] == 5.0)].count()['star_rating']
vine_yes_5star_perc = (vine_yes_5star / vine_yes_total) * 100

vine_yes_df = pd.DataFrame(
    [{"5 Star Ratings - Vine": vine_yes_5star,
     "Total Ratings - Vine": vine_yes_total,
     "Percentage 5 Star Ratings - Vine": vine_yes_5star_perc}])
vine_yes_df["Percentage 5 Star Ratings - Vine"] = vine_yes_df["Percentage 5 Star Ratings - Vine"].map("{:.2f}".format)
vine_yes_df
```

```python
vine_no_total = vine_no['review_id'].count()
vine_no_5star = vine_no[(vine_no['star_rating'] == 5.0)].count()['star_rating']
vine_no_5star_perc = (vine_no_5star / vine_no_total) * 100

vine_no_df = pd.DataFrame(
    [{"5 Star Ratings - No Vine": vine_no_5star,
     "Total Ratings - No Vine": vine_no_total,
     "Percentage 5 Star Ratings - No Vine": vine_no_5star_perc}])
vine_no_df["Percentage 5 Star Ratings - No Vine"] = vine_no_df["Percentage 5 Star Ratings - No Vine"].map("{:.2f}".format)
vine_no_df
```
